```python
def lexical(sentence, threshold=0.63, change=3):
  final = None
  if sentence[-1]=='?' or sentence[-1]=='!' or sentence[-1]=='.':
    final = sentence[-1]
    sentence =sentence[:-1]
  selected_tok = random.sample(sentence, min(change, len(sentence)))
  changed = False
  result = []

  for tok in sentence:
    if tok in selected_tok:
      try:
        sim_word, sim_score = model.most_similar(tok)[0]
        if sim_score > threshold:
            result.append(sim_word)
            changed = True
        else:
            result.append(tok)
      except Exception as e:
        result.append(tok)
    else:
        result.append(tok)
  if final:
    result.append(final)

  return result if changed else None
```

```python
def preprocess(sentence):
  sentence = re.sub(r'[^a-z0-9가-힣\.!\?\s]', '', sentence)
  return mecab.morphs(sentence)
```

```python
def build_corpus(df, thresholda=0.63, thresholdb=0.7):
  df.dropna(inplace=True)
  df['Q_']=df['Q'].apply(preprocess)
  df['A_']=df['A'].apply(preprocess)
  df.drop_duplicates(subset=['Q_'], inplace=True)
  df.drop_duplicates(subset=['A_'], inplace=True)
  df.reset_index(drop=True, inplace=True)

  que_corpus2 , ans_corpus2 = [],[]
  for i in range(len(df)):
    if len(df['Q_'][i])<28 and len(df['A_'][i])<35:
      num = random.choice([1, 2, 3])
      sim=lexical(df['A_'][i], change= num, threshold=thresholda)
      if sim is not None:
        que_corpus2.append(df['Q_'][i])
        ans_corpus2.append(['<SOS>']+sim+['<EOS>'])
      num = random.choice([5, 6, 7])
      sim=lexical(df['A_'][i], change= num, threshold=thresholda)
      if sim is not None:
        que_corpus2.append(df['Q_'][i])
        ans_corpus2.append(['<SOS>']+sim+['<EOS>'])

  df2=pd.DataFrame({'Q':que_corpus2, 'A':ans_corpus2})
  df2.drop_duplicates(subset=['A'], inplace=True)
  df2.reset_index(drop=True, inplace=True)
  que_corpus2=df2['Q'].tolist()
  ans_corpus2=df2['A'].tolist()

  que_corpus3 , ans_corpus3 = [],[]
  for i in range(len(df)):
    if len(df['Q_'][i])<28 and len(df['A_'][i])<35:
      num = random.choice([1, 2, 3])
      sim=lexical(df['Q_'][i], change= num, threshold=thresholdb)
      if sim is not None:
        que_corpus3.append(sim)
        ans_corpus3.append(['<SOS>']+df['A_'][i]+['<EOS>'])
      num = random.choice([5, 6, 7])
      sim=lexical(df['Q_'][i], change= num, threshold=thresholdb)
      if sim is not None:
        que_corpus3.append(sim)
        ans_corpus3.append(['<SOS>']+df['A_'][i]+['<EOS>'])

  df3=pd.DataFrame({'Q':que_corpus3, 'A':ans_corpus3})
  df3.drop_duplicates(subset=['Q'], inplace=True)
  df3.reset_index(drop=True, inplace=True)
  que_corpus3=df3['Q'].tolist()
  ans_corpus3=df3['A'].tolist()

  return que_corpus2+que_corpus3, ans_corpus2+ans_corpus3
```

```python
import itertools
```

```python
def lexical2(sentence, threshold=0.63, change=3):
  final = None
  if sentence[-1]=='?' or sentence[-1]=='!' or sentence[-1]=='.':
    final = sentence[-1]
    sentence =sentence[:-1]

  candidates=[]
  change=False
  for tok in sentence:
    try:
      candidate_temp = model.most_similar(tok)
      filtered = [w for w, score in candidate_temp if score >= threshold]
      if len(filtered)>=1:
        candidates.append(filtered)
        change=True
      else:
        candidates.append([])
    except Exception as e:
      candidates.append([])
  if not change:
    return None
  cand_idx = [i for i, c in enumerate(candidates) if len(c) >= 1]
  combi = list(itertools.combinations(cand_idx, min(change, len(sentence))))

  results=[]
  for i in combi:
    result=[]
    for idx, token in enumerate(sentence):
      if idx in i:
        result.append(candidates[idx][0])
      else:
        result.append(token)
    if final:
      result.append(final)
    results.append(result)
  return results
```

```python
from rouge_score import rouge_scorer

def compute_rouge_with_library(predictions, references):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    scores = {
        "rouge1": [],
        "rouge2": [],
        "rougeL": [],
    }

    for pred, ref in zip(predictions, references):
        result = scorer.score(ref, pred)
        for k in scores:
            scores[k].append(result[k].fmeasure)

    # 평균 F1 점수 계산
    averaged = {k: sum(v)/len(v) for k, v in scores.items()}
    return averaged['rouge1'], averaged['rouge2'], averaged['rougeL']

# 예시
preds = ["I love machine learning."]
refs  = ["I enjoy studying machine learning."]

print(compute_rouge_with_library(preds, refs))
```

```python
def calculate_bleu(reference, candidate, weights=[0.25, 0.25, 0.25, 0.25]):
    return sentence_bleu([reference],
                            candidate,
                            weights=weights,
                            smoothing_function=SmoothingFunction().method1)
```
